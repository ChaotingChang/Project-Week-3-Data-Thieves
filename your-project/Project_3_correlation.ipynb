```python
import pandas as pd
import numpy as np
```

```python
df_countries = pd.read_csv("/home/chao/Desktop/Ironhack/Projects/Project-Week-3-Data-Thieves/Data/countries.csv")
df_happiness = pd.read_csv("/home/chao/Desktop/Ironhack/Projects/Project-Week-3-Data-Thieves/Data/happiness_all.csv")
df_elections = pd.read_csv("/home/chao/Desktop/Ironhack/Projects/Project-Week-3-Data-Thieves/Data/Elections_10_countries.csv")
df_happiness = df_happiness[df_happiness["Year"] > 2010]
```

```python
happiness = pd.merge(df_happiness, df_countries, how ="left", left_on='id_country', right_on='id')
happiness = happiness.rename(columns = {"Year":"year"})
del happiness["id"]
```

```python
austria = happiness[happiness["country"] == "Austria"]
austria = austria.merge(df_elections, how="left", on=["id_country", "year"])
del austria["id_country"]
austria = austria.groupby("year").sum()
austria = austria.drop([2011, 2015]).reset_index()
austria = austria.replace(0,np.nan)
austria["proportion_votes"] = austria["proportion_votes"].bfill().ffill()
austria_corr = austria.corr(method="pearson")
austria_corr
#austria_result = austria_corr["proportion_votes"]
```
